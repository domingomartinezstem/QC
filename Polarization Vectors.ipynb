```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as axes3d
import matplotlib.gridspec as gridspec
```

```python
t = np.linspace(0, 20, 1000)
Hwave=np.cos(t+.75)  #horizontally polarized wave
Vwave=np.sin(t)  #veritically polarized wave

projected_angle=0
theta=projected_angle*(np.pi)/180
basis0=np.cos(theta)
basis1=np.sin(theta)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d', adjustable='box')
y1 = Hwave*basis0
z1 = Hwave*basis1
ax.plot(t, y1, z1,'g', label='curve')
y2 = Vwave*basis1
z2 = Vwave*basis0
ax.plot(t, y2, z2,'b')
supy=y1+y2
supz=z1+z2
ax.plot(t, supy, supz,'r')
ax.view_init(azim=20, elev=0) 
```

```python
#A 3D vector in python uses two points (x,y,z) and (dx,dy,dz)
#dx,dy, and dz represnt the change from the initial point 
#the format for 3D array is [x,y,z,dx,dy,dz]
Horz_vector = np.array([[0, 0, 0, 1, 0, 0],[0, 0, 0, -1, 0, 0]])
Vert_vector = np.array([[0, 2, 0, 0, 0, 1]])

X, Y, Z, U, V, W = zip(*Horz_vector)         #you may need to edit, copy, and paste this line
X1, Y1, Z1, U1, V1, W1 = zip(*Vert_vector)         

fig = plt.figure(figsize=(10,10))            #change figure size values for larger/smaller graph
ax = fig.add_subplot(111, projection='3d')   #do not edit this line 
ax.quiver(X, Y, Z, U, V, W)                  #you may need to edit, copy and paste this line 
ax.quiver(X1, Y1, Z1, U1, V1, W1)                 

ax.set_xlim([-2, 2])                         #change the axis values in the x-direction
ax.set_ylim([-2, 2])                         #change the axis values in the y-direction
ax.set_zlim([-2, 2])                         #change the axis values in the z-direction
ax.view_init(azim=40, elev=5)                #change these numbers to view the graph from different angles 
plt.show()                                   #do not edit this line
```
